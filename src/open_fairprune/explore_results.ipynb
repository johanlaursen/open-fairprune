```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np

json_file_path = "fairprune_hyperparameter_search.json"
df = pd.read_json(json_file_path)
df['prune_ratio'].unique()
df['prune_ratio'] = df['prune_ratio'].round(2)
df['beta'] = df['beta'].round(2)

# df[df['prune_ratio'] == 0]
```

```python
sns.lineplot(data=df, x="prune_ratio", y="EOdd", hue="beta", marker="o")
plt.xlabel("Prune Ratio", fontsize=16)
plt.ylabel("EOdd", fontsize=16)
plt.title("EOdd vs Prune Ratio", fontsize=16)
```

```python
sns.lineplot(data=df, x="prune_ratio", y="('matthews', 'total')", hue="beta", marker="o")
plt.xlabel("Prune Ratio", fontsize=16)
plt.ylabel("Matthews", fontsize=16)
plt.title("Matthews vs Prune Ratio", fontsize=16)
```

```python
df_2=df[df['prune_ratio'] == 0.3]
sns.lineplot(data=df_2, x="EOdd", y="('matthews', 'total')", marker="o")
plt.xlabel("EOdd", fontsize=16)
plt.ylabel("Matthews", fontsize=16)
plt.title("Performance vs fairness tradeoff (Prune Ratio = 0.3)", fontsize=16)
```
